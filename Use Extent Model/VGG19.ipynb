## 从VGG19中输出特征

```python
from tensorflow.keras.layers import *
from tensorflow.keras.utils import plot_model, to_categorical
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.applications import vgg16, vgg19
from tensorflow.keras.applications.vgg19 import preprocess_input


import matplotlib.pyplot as plt
import cv2
import numpy as np
```

## 获取VGG19模型

```python
base_model = vgg19.VGG19(weights="imagenet", include_top=False)
```

```python
image = plt.imread("005.jpg")    # 随便一张图，对形状没有要求
image = cv2.pyrDown(image)
image = np.expand_dims(image, axis=0)
image.shape
```

## 以VGG19的输入到名为“block4_pool”的层作为输出获得特征

```python
model = Model(inputs=base_model.input, outputs=base_model.get_layer('block4_pool').output)
```

## 获得指定图片的特征

```python
block4_pool_features = model.predict(image)
block4_pool_features.shape
```

```python
block4_pool_features.shape
```

```python
plt.figure(figsize=(10,10))

plt.subplot(3,3,2)
plt.imshow(image[0])
plt.title('origin image')

plt.subplot(3,3,4)
plt.imshow(block4_pool_features[0, :, :, 0])
plt.title('feature 1')

plt.subplot(3,3,5)
plt.imshow(block4_pool_features[0, :, :, 1])
plt.title('feature 2')

plt.subplot(3,3,6)
plt.imshow(block4_pool_features[0, :, :, 2])
plt.title('feature 3')

plt.subplot(3,3,7)
plt.imshow(block4_pool_features[0, :, :, 3])
plt.title('feature 4')

plt.subplot(3,3,8)
plt.imshow(block4_pool_features[0, :, :, 4])
plt.title('feature 5')

plt.subplot(3,3,9)
plt.imshow(block4_pool_features[0, :, :, 5])
plt.title('feature 6')
```
